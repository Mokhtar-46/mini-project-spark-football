```python
from pyspark.sql import SparkSession
from pyspark.ml import Pipeline
from pyspark.ml.feature import StringIndexer, VectorAssembler
from pyspark.ml.classification import RandomForestClassifier
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
import pandas as pd

# 1. Setup Spark session
spark = SparkSession.builder \
    .appName("SoccerMLTraining") \
    .getOrCreate()

print("--- Started AI model training process ---")

# 2. Load data
# Training data (historical)
train_path = "../outputs/prepared/cleaned_soccer_data.parquet"
# Test data (current season that we manually created)
test_path = "../data/current_season_scraped.csv"

print(f"📥 Loading training data from: {train_path}")
df_train = spark.read.parquet(train_path)

print(f"📥 Loading test data from: {test_path}")
df_test = spark.read.csv(test_path, header=True, inferSchema=True)

# 3. Prepare target variable
# Convert 'result' column (home_win, draw, away_win) to numbers (0, 1, 2)
# Because the machine does not understand text
def map_result(res):
    if res == 'home_win': return 0.0
    if res == 'draw': return 1.0
    if res == 'away_win': return 2.0
    return 0.0

# Apply the transformation (using UDF to ensure accuracy)
from pyspark.sql.functions import udf
from pyspark.sql.types import DoubleType

map_result_udf = udf(map_result, DoubleType())

df_train = df_train.withColumn("label", map_result_udf(col("result")))
# For test data, we will also need to convert results if available or use other features
# For simplicity, we assume we are testing the model's ability to predict results based on teams
df_test = df_test.withColumn("label", map_result_udf(col("result") if 'result' in df_test.columns else lit("draw")))

# 4. Feature Engineering
# Convert team names to numbers (StringIndexer)
indexer_home = StringIndexer(inputCol="home_team", outputCol="home_team_index")
indexer_away = StringIndexer(inputCol="away_team", outputCol="away_team_index")

# Assemble features into a single vector (VectorAssembler)
assembler = VectorAssembler(
    inputCols=["home_team_index", "away_team_index"],
    outputCol="features"
)

# 5. Build Pipeline
# Put all steps in a Pipeline to ensure any new data goes through the same processing stages
rf = RandomForestClassifier(labelCol="label", featuresCol="features", numTrees=100)

pipeline = Pipeline(stages=[indexer_home, indexer_away, assembler, rf])

# 6. Training
print("🚀 Training Random Forest model... (this may take a few moments)")
model = pipeline.fit(df_train)
print("✅ Training completed!")

# 7. Evaluation
print("🧪 Testing model on current season data...")
predictions = model.transform(df_test)

evaluator = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction", metricName="accuracy")
accuracy = evaluator.evaluate(predictions)

print(f"\n📊 Evaluation results:")
print(f"🎯 Model Accuracy: {accuracy * 100:.2f}%")

# 8. Display sample predictions
print("\n🔮 Sample of model predictions for upcoming matches:")
predictions.select("home_team", "away_team", "label", "prediction").show(10)

# 9. Save model (for future use)
model.save("../outputs/models/soccer_rf_model")
print("💾 Model saved in outputs/models/ folder")

spark.stop()
```
